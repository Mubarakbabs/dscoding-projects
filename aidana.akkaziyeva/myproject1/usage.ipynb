## Generating Cast Question

## Generating Country Question

```python
#filtering the dataframe: only movies with the values in the 'country' column not null 
non_empty = data[(data['type']=='Movie') & data['country'].notnull()] 
non_empty.head()
```

```python
show_row = non_empty.sample(n=1).iloc[0]  #choosing one row at random
show, correct_opt = show_row['title'], show_row['country'].split(',')[0].strip()    #the answer is chosen to be the first country only
show, correct_opt
```

```python
#extracting the 'country' column from the DataFrame
countries_column = non_empty['country']
#splitting each comma-separated string into a list of countries
#using a generator expression to iterate over each row's countries, splitting them by commas, and removing whitespace
#using `if country.strip()` to filter out empty strings
split_countries = (country.strip() for countries in countries_column for country in countries.split(',') if country.strip())
[split_countries][:5]

```

## Preparing Data for Visualisations

```python
#removing specific substrings from the 'duration' column
data['duration'] = data['duration'].str.replace(' Seasons', '')  # Remove ' Seasons' from the duration information
data['duration'] = data['duration'].str.replace(' min', '')      # Remove ' min' from the duration information
data['duration'] = data['duration'].str.replace(' Season', '')   # Remove ' Season' from the duration information

#converting the cleaned 'duration' column to numeric format
#errors='coerce' replaces any non-numeric values with NaN
data['duration'] = pd.to_numeric(data['duration'], errors='coerce')

data.head()
```

```python
#dataframe with movies that are less than 180 minutes
movies_data = data[(data['type'] == 'Movie') & (data['duration'] <= 180)]

movies_data.head()
```

```python
#dataframe with TV shows that have less than 11 seasons
tv_shows_data = data[(data['type'] == 'TV Show') & (data['duration'] <= 11)]
tv_shows_data.head()
```

```python
#counting the occurrences of each unique duration value and sorting in ascending order
movie_length = movies_data['duration'].value_counts().sort_index()
season_counts = tv_shows_data['duration'].value_counts().sort_index()
movie_length[:5], season_counts[:5]
```

```python
#creating a Series with release years as the index and their counts as values
show_release_year = data.loc[data['type'] == 'TV Show', 'release_year'].value_counts()
movie_release_year = data.loc[data['type'] == 'Movie', 'release_year'].value_counts()
show_release_year[:5], movie_release_year[:5]
```

```python
#combining TV show and movie release year counts into a single DataFrame
release_year_data = pd.DataFrame({'TV Shows': show_release_year, 'Movies': movie_release_year})
release_year_data.head()
```

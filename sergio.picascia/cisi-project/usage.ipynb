## Analysis

```python
import numpy as np
```

### Random Retrieval

```python
random_retrieval = {}

for query in data.queries['id'].unique():
    random_documents = np.random.choice(data.documents['id'], size=41)
    random_retrieval[query] = random_documents
```

```python
scores = []

for query in random_retrieval:
    y_pred = random_retrieval[query]
    y_true = data.relations.loc[data.relations['query'] == query, 'document']
    
    scores.append(len(np.intersect1d(y_pred, y_true)) / len(y_pred))
```

```python
np.array(scores).mean()
```

### Term Frequency

```python
vocabulary = set(' '.join(data.documents['text'].values).split(' '))
tf_documents = np.zeros((len(data.documents), len(vocabulary)))
vocabulary = dict(zip(vocabulary, range(len(vocabulary))))
```

```python
vocabulary.keys()
```

```python
tf_documents
```

```python
for i, txt in enumerate(data.documents['text'].values):
    for word in txt.split(' '):
        tf_documents[i][vocabulary[word]] += 1
        
for i, row in enumerate(tf_documents):
    tf_documents[i] = tf_documents[i]/np.sum(tf_documents[i])
```

```python
tf_similarity = np.empty((len(data.queries), tf_documents.shape[0]))
```

```python
for j, query in enumerate(data.queries['text'].values):
    q_words = list({vocabulary[word] for word in query.split(' ') if word in vocabulary})
    for i, document in enumerate(tf_documents):
        tf_similarity[j][i] = tf_documents[i][q_words].sum()
```

```python
tf_similarity
```

```python
results = tf_similarity.argsort(axis=1)[:, -41:]
```

```python
scores = []

for query in range(results.shape[0]):
    y_pred = results[query] + 1
    y_true = data.relations.loc[data.relations['query'] == query+1, 'document']
    
    scores.append(len(np.intersect1d(y_pred, y_true)) / len(y_pred))
```

```python
np.array(scores).mean()
```

```python
threshold = 0.5
scores = []

for i, query in enumerate(tf_similarity):
    y_pred = np.argwhere(query > threshold).flatten() + 1
    y_true = data.relations.loc[data.relations['query'] == i+1, 'document']
    
    if y_pred.any():
        scores.append(len(np.intersect1d(y_pred, y_true)) / len(y_pred))
    else:
        scores.append(0)
```

```python
np.array(scores).mean()
```

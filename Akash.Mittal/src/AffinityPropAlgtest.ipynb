```python
# Testing the AffinityPropagation algorithm on a sample dataset (using make_blobs function) to perform clustering.
# Reference - https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AffinityPropagation.html
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import AffinityPropagation
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=100, n_features=15, centers=3, cluster_std=1.0) #Generating Sample Data
```

```python
af = AffinityPropagation(damping = 0.9, preference = None) #Creating Affinity Object

# Note, when the affinitypropagation doesn't converge, we need to modify the damping parameter, otherwise without convergence the output of clustering is useless.
```

```python
af.fit(X)
```

```python
cluster_centers = af.cluster_centers_ #Get the Cluster Centers
```

```python
labels = af.labels_ #Get labels for each data point
```

```python
n_clusters = len(cluster_centers) # Number of Clusters
```

```python
# Now, plotting the clusters
```

```python
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
```

```python
for i in range(n_clusters):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], c=colors[i % len(colors)], label='Cluster {}'.format(i))

plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=100, color='black', label='Cluster centers')
plt.title('Affinity Propagation Clustering')
plt.legend()
plt.show()
```

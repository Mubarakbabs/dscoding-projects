```python
import pandas as pd
```

```python
df = pd.read_csv("bgg.csv")
df
```

```python
df_sorted = df.sort_values(['title', 'rating'], ascending=(True, True))
df_sorted
```

```python
dftitle=df_sorted['title'].value_counts()
dftitle
```

```python
total_rating = dftitle.sum()
total_rating
```

```python
nan_counts_per_movie = df_sorted.groupby('title')['rating'].apply(lambda x: x.isna().sum()).reset_index()
nan_counts_per_movie
```

```python
nan_sum=nan_counts_per_movie['rating'].sum()
nan_sum
```

```python
non_nan_counts_per_movie = df_sorted.groupby('title')['rating'].apply(lambda x: x.notna().sum()).reset_index()
non_nan_counts_per_movie
```

```python
non_nan_sum=non_nan_counts_per_movie['rating'].sum()
non_nan_sum
```

```python
if non_nan_sum+nan_sum == total_rating:
    print("verified")
else:
    print("incorrect")
```

```python
highest_rating_per_movie = df_sorted.groupby('title')['rating'].max().reset_index()
highest_rating_per_movie
```

```python
lowest_rating_per_movie = df_sorted.groupby('title')['rating'].min().reset_index()
lowest_rating_per_movie
```

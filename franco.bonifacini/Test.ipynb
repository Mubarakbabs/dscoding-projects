```python
def Hotel_list():
    hotel_list = []
    for hotel, rooms in zip(hotels_db['hotel'], hotels_db['rooms']):
        hotel_list.extend([hotel] * rooms)
    return hotel_list
```

```python
def Random_guest():
    Random_guests = list(np.random.choice(guests_db["guest"],size = len(guests_db), replace = False))
    return Random_guests
```

```python
#Random_guest()
```

```python
def Random_hotel():
    hotel_list = Hotel_list()
    np.random.shuffle(hotel_list)
    return hotel_list
```

```python
#Random_hotel()
```

```python
def Random_allocation():
    guest = Random_guest()
    hotel = Random_hotel()
    length = min(len(guest),len(hotel))
    allocation = [(guest, hotel) for guest, hotel in zip(guest[:length],hotel[:length])]
    return allocation
```

```python
def Order_by_price():
    price_sort = hotels_db.sort_values(by=['price'])
    hotel_list = []
    for hotel, rooms in zip(price_sort['hotel'], price_sort['rooms']):
        hotel_list.extend([hotel] * rooms)
    return hotel_list
```

```python
def Order_by_qrooms():
    rooms_sort = hotels_db.sort_values(by=['rooms'], ascending=False)
    hotel_list = []
    for hotel, rooms in zip(rooms_sort['hotel'], rooms_sort['rooms']):
        hotel_list.extend([hotel] * rooms)
    return hotel_list
```

```python
def Random_room():
    hotel_occupancy = {hotel: list(range(1, rooms+1)) for hotel, rooms in zip(hotels_db['hotel'], hotels_db['rooms'])}
    hotel_order = list(hotel_occupancy.keys())
    np.random.shuffle(hotel_order)
    random_rooms = []
    while any(hotel_occupancy.values()):
        for hotel in hotel_order:
            available_rooms = hotel_occupancy[hotel]
            if available_rooms:
                selected_room = np.random.choice(available_rooms)
                random_rooms.append((hotel, selected_room))
                available_rooms.remove(selected_room)
    random_room = [(hotel, room) for hotel, room in random_rooms]
    return random_room
```

```python
def Assigned_hotel():
    hotel_occupancy = {hotel: list(range(1, rooms + 1)) for hotel, rooms in zip(hotels_db['hotel'], hotels_db['rooms'])}
    assigned_hotels = [hotel for hotel, room_numbers in hotel_occupancy.items() for _ in room_numbers]
    return assigned_hotels
```

```python
def Random_allocation():
    guest = Random_guest()
    hotel = Assigned_hotel()
    length = min(len(guest),len(hotel))
    allocation = [[guest, hotel] for guest, hotel in zip(guest[:length],hotel[:length])]
    return allocation
```
